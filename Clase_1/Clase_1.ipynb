# Hacer un primer gráfico con matplotlib y seaborn
[Este es la presentación](https://docs.google.com/presentation/d/1OhwYkGXfY9xDGjP7cRscmms6Cd4ogESgaRvRY17fY-I/edit?slide=id.g221baacdfc3_0_152#slide=id.g221baacdfc3_0_152)

```python
import pandas
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
x:list[str] = ['EDA','Data Wrangling','Mineria datos','Limpieza Datos','SQL','Est. Inferencial','ML','DL','IA']
y:list[int] = [5,10,15,20,30,25,31,30,20]
```

```python
data_dict:dict = {
    "Características": x,
    "Frecuencia": y
    }

data_dict
```

```python
df = pandas.DataFrame(data_dict) #dataframe
```

```python
df
```

```python
from matplotlib import pyplot as plt #codigo de inteligencia artificial no el esperado
df['Frecuencia'].plot(kind='hist', bins=20, title='Frecuencia')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
plt.figure(figsize=(16,9))
sn.barplot(df, x="Características", y="Frecuencia")
plt.show()
```
