# Clase 5 Visualizaciones avanzadas

```python
# Generar data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,200)
y = np.tanh(x)

data = pd.DataFrame({'x':x,'y':y})
data.head()
```

```python
# que y donde voy a colocoar esos datos
plt.figure(figsize=(8,4))

# los datos y que voy a colocar
plt.plot(data["x"], data["y"], label= "y = tanh(x)", color = (0.1, 0.2, 0.5, 0.3), linestyle= ":")
plt.title("Gr´pafico de función tangencial")

# leyendas, información adicional al gráfico
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")

plt.text(5,0.5,r"$\sigma(z) = \frac{e^z - e^{-z}}")

plt.grid(True, alpha = 0.5)
plt.tight_layout( pad = 2)
plt.legend()
plt.show()
```

Cuadratica

```python
# Generar datos de entrada

x = np.linspace(-10, 10, 200)  # 200 valores entre -10 y 10

# Definir una relación cuadrática: y = ax^2 + bx + c

a, b, c = 2, -3, 5

y = a * x**2 + b * x + c

cuadradica = pd.DataFrame({'x':x,'y':y})

cuadradica.head()

#Visualizar gráfica
plt.figure(figsize=(8, 5))

plt.plot(x, y, label=f"${a}x^2 + {b}x + {c}$")
plt.title("Curva Cuadrática")
plt.xlabel("x")
plt.ylabel("f(y)")
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```

```python
np.random.seed(42)
x = np.random.rand(50)
y = 2 * x + 1 + np.random.normal(scale = 0.1, size = 50)

data = pd.DataFrame({"x":x, "y":y})
data.head()
```

```python
football_csv_path = "https://raw.githubusercontent.com/Andru-1987/74235-_DataScience_I/refs/heads/main/clase_5/storage/futbolargentino(Worksheet).csv"
```

```python
football_dataset = pd.read_csv(football_csv_path, sep=";")
```

```python
football_dataset.info()
```

```python
plt.figure(figsize=(8,5))
plt.hist(football_dataset.Edad, bins = 10, color ="skyblue", edgecolor="black")

#etiquetas y titulos
plt.title("Distribución de dades de jugadores de futbol")
plt.xlabel("Edad")
plt.ylabel("Cantidad de jugadores")
plt.grid(axis = "y", linestyle ="--", alpha = 0.7)
plt.tight_layout()
plt.show()
```

```python
football_dataset.info()
```

```python
football_dataset["edad_cleaned"] = football_dataset["Edad"].astype(str).str.extract(r"(\d+)").astype(int)
```

```python
football_dataset.info()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(football_dataset["edad_cleaned"], vert= False, patch_artist=True,
            boxprops= dict(facecolor ="skyblue", color="black"),
            medianprops = dict(color ="red"),
            whiskerprops= dict(color ="black"),
            capprops = dict(color ="black"),
            flierprops=dict(markerfacecolor="gray", marker="o", markersize = 5, linestyle ="none", markeredgecolor="black"))


#etiquetas y titulos
plt.title("Boxplot edades de jugadores de futbol")
plt.xlabel("Edad")
plt.grid(axis="x", linestyle ="--", alpha = 0.7)
plt.tight_layout()
plt.show()
```

```python
football_dataset_normalised = football_dataset[ football_dataset["edad_cleaned"] <= 38.5]
```

```python
football_dataset_normalised
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(football_dataset_normalised["edad_cleaned"], vert=False, patch_artist = True,
            boxprops=dict(facecolor="skyblue", color="black"),
            medianprops=dict(color="red"),
            whiskerprops=dict(color="black"),
            capprops=dict(color="black"),
            flierprops=dict(markerfacecolor="gray", marker="o", markersize=5, linestyle="none", markeredgecolor="black")
            )

#Etiquetas y título
plt.title("Boxplot de edades de jugadores de fútbol")
plt.xlabel("Edad")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(football_dataset_normalised["edad_cleaned"], bins=20, color='skyblue', edgecolor='black')

#Etiquetas y título
plt.title("Boxplot de edades de jugadores de fútbol")
plt.xlabel("Edad")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

Gráficos en 3d

```python
#simulamos un dataset realista
np.random.seed(42)
n = 100

data = {
    'Ventas': np.random.randint(50_000,75_000, n),
    'Satisfaccion_Cliente': np.random.uniform(1, 10, n),  # puntaje 1 a 5
    'Costo_Operativo': np.random.randint(20_000, 40_000, n),  # costo entre 500 y 4000
    'Region': np.random.choice(['Norte', 'Sur', 'Este', 'Oeste'], n),
    'Producto': np.random.choice(['Tv+Cable', 'Internet', 'Pack Football'], n)
}

data_internet = pd.DataFrame(data)
data_internet.head()
```

```python
#mapear color por región
colores = {'Norte': 'red', 'Sur': 'blue', 'Este': 'green', 'Oeste': 'purple'}
colors = data_internet['Region'].map(colores)
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")

scatter_plot = ax.scatter(
        data_internet['Ventas'],
        data_internet['Satisfaccion_Cliente'],
        data_internet['Costo_Operativo'],
        c=colors,
        alpha=0.7,
        label=data_internet["Region"]
        )

#Etiquetas
ax.set_xlabel("Ventas")
ax.set_ylabel("Satisfacción del Cliente")
ax.set_zlabel("Costo Operativo")
ax.set_title("Ventas vs Satisfacción del Cliente vs Costo Operativo")

#Leyenda personalizada
from matplotlib.lines import Line2D
legend_elements =[Line2D([0], [0], marker="o", color="w", label=reg,
                         markerfacecolor=col, markersize=10) for reg, col in colores.items()]

ax.legend(handles=legend_elements, title="Región")

plt.show()
```
